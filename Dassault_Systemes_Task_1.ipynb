```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')
```

```python
# Path to the CSV file in Google Drive
file_path = '/content/drive/My Drive/Colab Notebooks/healthcare_dataset.csv'

# Loaded  the data into a pandas DataFrame
df = pd.read_csv(file_path)
```

```python

# a. Create a plot to find out if there is any correlation between Blood Group Type, Gender, and Medical Condition
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Medical Condition', hue='Blood Group Type', palette='Set2')
plt.title('Blood Group Type distribution among different Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Blood Group Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Medical Condition', hue='Gender', palette='Set1')
plt.title('Gender distribution among different Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
```

```python

# b. Find out the Average Billing Amount required to treat each medical Condition
average_billing = df.groupby('Medical Condition')['Billing Amount'].mean().reset_index()
print(average_billing)
```
